```python
# Plot the given dictionary such that all the serires are in one plot
import matplotlib.pyplot as plt
```

```python
def plot(arr45, arr90, arr135, arr180, L, title):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)

    vals = range(L)
    x_array = [value * (180/L) for value in vals]

    ax1.scatter(x_array, arr45, marker="s", label='45')
    ax1.scatter(x_array, arr90, marker="o", label='90')
    ax1.scatter(x_array, arr135, marker="v", label='135')
    ax1.scatter(x_array, arr180, marker="x", label="180")

    plt.title(title)

    plt.legend()
    plt.show()

    plt.savefig("{}.png".format(title))
```
